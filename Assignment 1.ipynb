### Converting categorical variables into quantitative variables

```python
dictionary = {'OUTLOOK': ['Rainy', 'Rainy',
                          'Overcast', 'Sunny',
                          'Sunny', 'Sunny',
                          'Overcast', 'Rainy',
                          'Rainy', 'Sunny',
                          'Rainy', 'Overcast',
                          'Overcast', 'Sunny'],
              'TEMPERATURE': ['Hot', 'Hot', 'Hot',
                              'Mild', 'Cool',
                              'Cool', 'Cool',
                              'Mild', 'Cool',
                              'Mild', 'Mild',
                              'Mild', 'Hot', 'Mild']}
  
# converting the dictionary to DataFrame
df2 = pd.DataFrame(dictionary)
display(df2)
```

```python
pd.get_dummies(df2['TEMPERATURE'])
```

#### Dummy Variables act as indicators of the presence or absence of a category in a Categorical Variable. 0 represents absence while 1 represents presence.
